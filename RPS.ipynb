```python
import random

options = ["rock", "paper", "scissors"]

while True:
    try:
        user_choice = input("You choose. Rock, Paper, Scissors: ").lower().strip()
        if user_choice == 'rock':
            break
        elif user_choice == 'paper':
            break
        elif user_choice == 'scissors':
            break   
        else:
           print("Uh oh. Try Again.")
    except ValueError:
        print("Uh oh. Try again.")
    
user_choice = input("You choose. Rock, Paper, Scissors: ").lower().strip()

if user_choice == 'rock':
    print('You: rock')
elif user_choice == 'paper':
    print('You: paper')
elif user_choice == 'scissors':
    print('You: scissors')


      
computer_choice = random.choice(options)
print(f"Bot Man: {computer_choice}")


if user_choice == computer_choice:
    print("It's a tie!")
elif user_choice == 'rock' and computer_choice == 'scissors':
    print("You Win!")
elif user_choice == 'scissors' and computer_choice == 'paper':
    print("You Win!")
elif user_choice == 'paper' and computer_choice == 'rock':
    print("You Win!")    
else:
    print("Nice Try! Try again.")




```
